**Setup**

```python
from pycaret.time_series import *
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])

df.set_index('DATE', inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
s = setup(df, target='IPG2211A2N',fold=2,fh=12)
```

```python
check_stats()
```

**Create Model**

```python
model = create_model('naive')
```

```python
finalized_model = finalize_model(model)
```

```python
finalize_model
```

**Predictions**

```python
prediction = predict_model(model)
```
